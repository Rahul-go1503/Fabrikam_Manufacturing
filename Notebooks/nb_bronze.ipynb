```python
from pyspark.sql.functions import current_timestamp, input_file_name

# File path (uploaded to Lakehouse Files or mounted location)
file_path = "Files/crm_customer_profiles.csv"

# Read CSV into DataFrame (raw ingestion)
df_bronze = (
    spark.read
    .format("csv")
    .option("header", "true")
    .load(file_path)
    # Add metadata columns
    .withColumn("ingestion_timestamp", current_timestamp())
    .withColumn("source_file", input_file_name())
)

# Save to Bronze table
df_bronze.write.format("delta").mode("overwrite").saveAsTable("bronze_crm_customer_profiles")
```
